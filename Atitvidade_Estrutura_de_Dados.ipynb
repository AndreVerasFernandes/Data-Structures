Estruturas ligadas e Filas

Classe Node: Esta classe representa uma solicitação de impressão individual. Cada nó tem um name (nome do solicitante), company (empresa do solicitante) e next (a próxima solicitação na fila).

Classe Queue: Esta classe representa a fila de impressão para uma impressora específica. Ela tem dois atributos principais: head (a primeira solicitação na fila) e tail (a última solicitação na fila). A classe Queue tem quatro métodos principais:

enqueue(name, company): Adiciona uma nova solicitação à fila.

dequeue(): Remove e retorna a primeira solicitação da fila.

is_empty(): Verifica se a fila está vazia.

print_queue(): Imprime todas as solicitações na fila.

Funções de Gerenciamento de Impressão: O código define várias funções para gerenciar as filas de impressão:

process_print_request(name, company, floor): Processa uma solicitação de impressão, adicionando-a à fila de impressão apropriada com base no andar fornecido.

check_queues(): Verifica o estado atual de todas as filas de impressão.

print_process(): Simula o processo de impressão, imprimindo todas as solicitações em todas as filas até que todas as filas estejam vazias.


```python
class Node:
    def __init__(self, name, company):
        self.name = name
        self.company = company
        self.next = None

class Queue:
    def __init__(self):
        self.head = None
        self.tail = None

    def enqueue(self, name, company):
        new_node = Node(name, company)
        if not self.head:
            self.head = new_node
            self.tail = new_node
        else:
            self.tail.next = new_node
            self.tail = new_node

    def dequeue(self):
        if not self.head:
            return None
        else:
            name = self.head.name
            company = self.head.company
            self.head = self.head.next
            if not self.head:
                self.tail = None
            return name, company

    def is_empty(self):
        return self.head is None

    def print_queue(self):
        current = self.head
        while current:
            print(f"Nome: {current.name}, Empresa: {current.company}")
            current = current.next
        print("")


floor1_printer_queue = Queue()
floor2_printer_queue = Queue()
floor3_printer_queue = Queue()


def process_print_request(name, company, floor):
    if floor == 1:
        floor1_printer_queue.enqueue(name, company)
        print(f"{name} da empresa {company} adicionado à fila do andar 1")
    elif floor == 2:
        floor2_printer_queue.enqueue(name, company)
        print(f"{name} da empresa {company} adicionado à fila do andar 2")
    elif floor == 3:
        floor3_printer_queue.enqueue(name, company)
        print(f"{name} da empresa {company} adicionado à fila do andar 3")


def check_queues():
    print("Fila do andar 1:")
    floor1_printer_queue.print_queue()
    print("Fila do andar 2:")
    floor2_printer_queue.print_queue()
    print("Fila do andar 3:")
    floor3_printer_queue.print_queue()

def print_process():
    while not (floor1_printer_queue.is_empty() and floor2_printer_queue.is_empty() and floor3_printer_queue.is_empty()):
        if not floor1_printer_queue.is_empty():
            name, company = floor1_printer_queue.dequeue()
            print(f"Imprimindo {name} da empresa {company} do andar 1")
        if not floor2_printer_queue.is_empty():
            name, company = floor2_printer_queue.dequeue()
            print(f"Imprimindo {name} da empresa {company} do andar 2")
        if not floor3_printer_queue.is_empty():
            name, company = floor3_printer_queue.dequeue()
            print(f"Imprimindo {name} da empresa {company} do andar 3")

    print("Todos os documentos foram impressos.")


process_print_request("Ruan", "Soluções Luminar", 1)
process_print_request("Andre", "Inovações Nexus", 2)
process_print_request("Giovanna", "Estúdio Criativo Arco-Íris", 3)
process_print_request("Ana", "Soluções Luminar", 1)


print("Estado das filas antes da impressão:")
check_queues()


print("\nIniciando processo de impressão:")
print_process()


print("\nEstado das filas após a impressão:")
check_queues()
```

## Pilhas

As funções de desfazer (undo, CTRL + Z) e refazer (redo, CTRL + Y) são recursos essenciais em muitos sistemas de edição de texto, permitindo aos usuários reverter ou repetir suas ações.  Essas funções são implementadas através do uso de duas pilhas: uma para desfazer e outra para refazer.

No código Python, a classe EditorDeTexto implementa essas funções. Aqui está uma explicação detalhada:

Quando o usuário digita ou deleta texto, o estado atual do texto é adicionado à pilha de desfazer (undo_stack). O texto é então atualizado e a pilha de refazer (redo_stack) é limpa.

A função desfazer verifica se a pilha de desfazer não está vazia. Se não estiver, o estado atual do texto é adicionado à pilha de refazer, e o último estado do texto na pilha de desfazer é recuperado e se torna o estado atual do texto.

A função refazer funciona de maneira semelhante à função desfazer, mas ao contrário. Verifica se a pilha de refazer não está vazia e, se não estiver, o estado atual do texto é adicionado à pilha de desfazer, e o último estado do texto na pilha de refazer é recuperado e se torna o estado atual do texto.

A função obter_texto simplesmente retorna o estado atual do texto.

Este código demonstra um exemplo de como as funções de desfazer e refazer podem ser implementadas em um editor de texto simples. No entanto, em aplicativos do mundo real, essas funções podem ser muito mais complexas, lidando com a formatação de texto, inserção/deleção de imagens, e assim por diante.




```python
class EditorDeTexto:
    def __init__(self):
        self.texto = ""
        self.undo_stack = []
        self.redo_stack = []


    def digitar(self, texto):
        self.undo_stack.append(self.texto)
        self.texto += texto
        self.redo_stack.clear()

    def deletar(self, num_caracteres=1):

        self.undo_stack.append(self.texto)

        self.texto = self.texto[:-num_caracteres]
        self.redo_stack.clear()

    def desfazer(self):
        if self.undo_stack:

            self.redo_stack.append(self.texto)
            self.texto = self.undo_stack.pop()

    def refazer(self):
        if self.redo_stack:

            self.undo_stack.append(self.texto)
            self.texto = self.redo_stack.pop()

    def obter_texto(self):
        return self.texto

    def salvar_arquivo(self, nome_arquivo):
        with open(nome_arquivo, "w") as arquivo:
            arquivo.write(self.texto)

editor = EditorDeTexto()
```

```python
editor.digitar("Olá, mundo! ")
editor.digitar("Isso é um exemplo.")
print(editor.obter_texto())
```

```python
editor.deletar()
print(editor.obter_texto())
```

```python
editor.desfazer()
print(editor.obter_texto())
```

```python
editor.refazer()
print(editor.obter_texto())
```

```python
editor.deletar(5)
print(editor.obter_texto())
```

```python
editor.deletar(50)
print(editor.obter_texto())
editor.salvar_arquivo("arquivo.txt")
```

```python
class NotasAlunos:
    def __init__(self, notas):
        self.notas = notas

    def get_notas(self):
        return self.notas

    def inserir_nota(self, index, aluno, nota):
        self.notas.insert(index, (aluno, nota))
        return (aluno, nota) in self.notas

    def remover_nota(self, aluno, nota):
        self.notas.remove((aluno, nota))
        return (aluno, nota) not in self.notas

    def notas_procurar(self, nota_minima):
        nomes = []
        for aluno, nota in self.notas:
            if nota >= nota_minima:
                nomes.append(aluno)
        return nomes

    def tamanho_do_array(self):
        return len(self.notas)


notas_exemplo = [('Ana', 6.7), ('Carlos', 8.5), ('João', 7.5), ('Maria', 8.3), ('Pedro', 9.0)]
manipulador = NotasAlunos(notas_exemplo)

print("Notas originais:", manipulador.get_notas())

manipulador.inserir_nota(4, 'Lucas', 10)
print("Notas após inserção:", manipulador.get_notas())

manipulador.remover_nota('Ana', 6.7)
print("Notas após remoção:", manipulador.get_notas())

nota_minima_desejada = 8.0
nomes_notas_maiores = manipulador.notas_procurar(nota_minima_desejada)
print(f"Nomes dos alunos com notas maiores ou iguais a {nota_minima_desejada}: {nomes_notas_maiores}")

tamanho_array = manipulador.tamanho_do_array()
print(f"Quantidade de notas: {tamanho_array}")
```
